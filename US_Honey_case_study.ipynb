# Find out which state was most contributing to make the year 2000 as the max producer.

```python
all_states = df.groupby(['state','year'])['production'].sum().reset_index()
```

```python
all_states
```

```python
# Now we are going to select all the rows where year =2000
top_state_2000 = all_states.loc[(all_states['year']==2000)].reset_index()
```

```python
top_state_2000.sort_values(by='production',ascending=False).head(1)
top_state_2000
```

 # 1- Create two barplots for these top_5_states and bottom_5_states

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create barplot for top 5 states
plt.figure(figsize=(10,6))
sns.barplot(x=top_5_states.index, y=top_5_states.values)
plt.title('Top 5 States with Highest Production')
plt.xlabel('State')
plt.ylabel('Production')
plt.show()
```

```python
# Create barplot for bottom 5 states
plt.figure(figsize=(10,6))
sns.barplot(x=bottom_5_states.index, y=bottom_5_states.values)
plt.title('Bottom 5 States with Lowest Production')
plt.xlabel('State')
plt.ylabel('Production')
plt.show()
```

# 2- Which sates has the highest number of colonies in the year 2000.

```python
colonies_2000 = df[df['year'] == 2000]
max_colonies_state = colonies_2000.loc[colonies_2000['colonies_number'].idxmax(), 'state']
print(f"The state with the highest number of colonies in 2000 is: {max_colonies_state}")
```

# 3- Create a heat map to check the relation among the columns.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
correlation_matrix = numeric_df.corr()
```

```python
# Plot the heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# 4- Find out how much the change in mean average price of honey from 1995 to 2021.

```python
mean_price_1995 = df[df['year'] == 1995]['average_price'].mean()
#mean_price_1995 = df[df['year'] == 1995]['price_per_lb'].mean()
```

```python
mean_price_2021 = df[df['year'] == 2021]['average_price'].mean()
```

```python
change_in_mean_price = mean_price_2021 - mean_price_1995
```

```python
print(f"The change in mean average price of honey from 1995 to 2021 is: {change_in_mean_price}")
```
